## Control Lyapunov Functions (CLF)

::: {.callout-note title="Definition: Control Lyapunov Function (CLF)"}
Consider a nonlinear system
$$
\dot{x} = F(x, u)
$$
A continuously differentiable function $V: D \to \mathbb{R}$, where $0 \in D$, is an **asymptotically stabilizing control Lyapunov function (CLF)** for $\dot{x} = F(x, u)$ if:

* $V(0) = 0$, and $V(x) > 0, \quad \forall x \in D \text{ with } x \ne 0$ (i.e., $V$ is positive definite).
* $\exists u \in U$, such that $\dot{V}(x, u) < 0, \quad \forall x \in D \text{ with } x \ne 0$.

This second condition is equivalent to:
$$
\inf_{u \in U} \dot{V}(x, u) < 0
$$
:::

To make it exponentially stabilizing CLF another constraint is needed and some needed to be altered: 

* $k_1\|x\|^a \le V(x) \le k_2\|x\|^a$
* $\exists u \in U$, s.t. $\dot{V}(x, u) + k_3V(x) \le 0, \quad \forall x \in D \text{ with } x \ne 0$.

This second condition is equivalent to:
$$
\inf_{u \in U} \left[ \dot{V}(x, u) + k_3V(x) \right] \le 0
$$

::: {.callout-tip title="Example: Stabilization of a Linear System via CLF"}
Consider a linear system:
$$
\dot{x} = Ax + Bu.
$$
Choose a quadratic Lyapunov function:
$$
V(x) = x^T Px, \quad P > 0.
$$
The CLF condition requires finding a control $u$ such that the time derivative $\dot{V}(x)$ satisfies the exponential decay criterion:
$$
{V}(x)=\frac{\delta{V}}{(\delta x)}\dot{x}
$$
$$
\dot{V}(x) = x^T(PA + A^T P)x + 2x^T PBu \le -c_3V(x).
$$
This condition must hold for some positive constant $c_3$. The inequality helps define the minimum required control effort $u$.
:::

In control affine system the equation looks something like this:

Consider a nonlinear control system in the **control-affine form**:
$$
\dot{x} = \underbrace{f(x)}_{\text{drift term / autonomous vector field}} + \underbrace{g(x)u}_{\text{actuation effect / control vector field}}
$$
where $f: \mathbb{R}^n \to \mathbb{R}^n$ and $g: \mathbb{R}^n \to \mathbb{R}^{n \times m}$ are **locally Lipschitz continuous in $x$**.

::: {.callout-note title="Definition (CLF for Control Affine Systems)"}
A continuously differentiable function $V: \mathbb{R}^n \to \mathbb{R}$ is an **exponentially stabilizing CLF** for
$$
\dot{x} = f(x) + g(x)u
$$
if there exist constants $c_1, c_2, c_3 > 0$ such that:

$$
c_1\|x\|^2 \le V(x) \le c_2\|x\|^2,
$$
and
$$
\inf_{u \in U} \left[ L_f V(x) + L_g V(x)u + c_3V(x) \right] \le 0, \quad \forall x \in \mathbb{R}^n.
$$
:::

## ðŸ“ Quadratic Program (QP) Overview

* A **Quadratic Program (QP)** solves the optimization problem:
    $$
    \min_{x \in \mathbb{R}^n} \frac{1}{2} x^T Qx + c^T x \quad \text{subject to } Ax \le b
    $$

* The objective is quadratic in $x$ and the constraints are linear inequalities, where:
    * $Q \in \mathbb{R}^{n \times n}$: **symmetric, positive semidefinite matrix**
    * $c \in \mathbb{R}^n$: **linear term**
    * $A \in \mathbb{R}^{m \times n}, b \in \mathbb{R}^m$: **inequality constraints**

* The QP is **convex** if $Q \succeq 0$ (positive semidefinite), ensuring a **unique global optimum**.

* Additional constraints like equalities and bounds can be included.

* QPs can be solved efficiently using MATLAB QP Solvers or Optimization Toolboxes, e.g., `quadprog`, `fmincon`, CVX.

## Safety

::: {.callout-note title="Definition (Safety)"}
Let $X$ be the state space and $X_u \subset X$ be the **unsafe set**. The system $\dot{x} = f(x)$ is **safe with respect to set $S \subset X$** if:

* $S \cap X_u = \emptyset$, and
* it is forward invariant with respect to $\dot{x} = f(x)$.

::: {.callout-note title="Nagumo's Theorem"}
Consider the safe set $S := \{x \in \mathbb{R}^n \mid h(x) \ge 0\}$.

* **Nagumo's Theorem:** If $h(x(0)) \ge 0$ (the system starts safe) and
    $$
    \dot{h}(x) \ge 0 \quad \text{for all } x \in \partial S,
    $$
    then $h(x(t)) \ge 0$ for all $t \ge 0$; that is, the safe set $S$ is **forward invariant**.

* **Interpretation:** This theorem certifies that if the system starts safe, then it remains safe.

::: {.callout-tip title="Key Comparison: Safety vs. Stability"}

| Safety (CBFs) | Stability (CLFs) |
| :--- | :--- |
| Enforced via **CBFs** (Control Barrier Functions). | Enforced via **CLFs** (Control Lyapunov Functions). |
| Ensures **forward invariance** of a safe set $C = \{x: h(x) \ge 0\}$. | Ensures **convergence** to a desired equilibrium (e.g., $x=0$). |
| "Bad things" (e.g., collisions) **never occur**. | "Good things" **eventually happen**. |

---

### Key Comparison

* **CBF:** Guarantees the system never leaves $C$.
* **CLF:** Guarantees the system converges to a target point.
