# Network systems

## Graph theory

!["Graph theory"](../images/Screenshot%202025-12-18%20at%2013.49.01.png)

A graph is weighted if each edge has an associated weight. A Parallel edges are two or more edges linking the same (ordered) pair of nodes. Self-loops are edges taht starts and ends at the same node. A graph is simple as it does not have parallel edges and self-loops. This course only considers simple graphs.
A undirected graph is connected if there is a path from i to j, for all i,j in V. It is an undirected graph when the edges do not have a direction.

::: {.callout-note title="3 levels of connection"}
 **Strongly connected**: if each node can reach all the other nodes via directed paths.
 **quasi-strongly connected** if there is a node (called *root node*) that can reach all the other nodes via directed paths
 **Weakly connected** If the underlying undirected graphs is connected


:::

!["Adjacency matrix"](../images/Screenshot%202025-12-18%20at%2014.05.00.png)

In this image it is also shown that when it is a weighted graph that the number should be the weight. If no weight is given the number should be 1. If a matrix is symmetric that also means that it is a undirected graph. This is only for simple graphs. A diagonal degree matrix D shows how many edges there are from node *i* with their weights so it is the sum of the weights of the edges from node *i*.

Incidence matrix are different these show which node is connected by what edge.
